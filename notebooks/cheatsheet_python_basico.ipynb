## 2. Operadores

```python
# Operadores aritméticos y de comparación
a = 10
b = 3

resultado = {
    "suma": a + b,
    "resta": a - b,
    "multiplicacion": a * b,
    "division": a / b,
    "division_entera": a // b,
    "modulo": a % b,
    "potencia": a ** b,
    "a_mayor_b": a > b
}
resultado
```

## 3. Condicionales

```python
edad = 20

if edad >= 18:
    print("Mayor de edad")
elif edad > 12:
    print("Adolescente")
else:
    print("Niño")
```

## 4. Bucles

```python
# Bucle for
for i in range(5):
    print(i)
```

```python
# Bucle while
contador = 0
while contador < 3:
    print("Hola")
    contador += 1
```

## 9. Ejercicios típicos resueltos

### Ejercicio 1: Número par o impar

```python
num = 7
if num % 2 == 0:
    print("Par")
else:
    print("Impar")
```

### Ejercicio 2: Mayor de tres números

```python
a, b, c = 10, 20, 15
mayor = max(a, b, c)
mayor
```

### Ejercicio 3: Sumar elementos de una lista con función

```python
def suma_lista(lista):
    return sum(lista)

suma_lista([1, 2, 3, 4])
```

### Ejercicio 4: Números del 1 al 10 con while

```python
i = 1
while i <= 10:
    print(i)
    i += 1
```
