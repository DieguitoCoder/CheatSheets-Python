```python
# EJERCICIOS 1-4: VARIABLES, INPUT Y OPERACIONES

# Ejercicio 1: Variables
print("--- Ejercicio 1 ---")
nombre = "Tu Nombre Completo"
edad = 25
print(f"1. Nombre: {nombre}, Edad: {edad}")

# Ejercicio 2: Input y Conversión
print("\n--- Ejercicio 2 ---")
try:
    # input() lee como string, int() convierte a entero
    num_a = int(input("2. Ingresa el primer número: "))
    num_b = int(input("2. Ingresa el segundo número: "))
    print(f"   Números ingresados: {num_a} y {num_b}")
except ValueError:
    print("   Por favor, ingresa un número entero válido.")


# Ejercicio 3: Operadores (Área de un Círculo)
print("\n--- Ejercicio 3 ---")
pi = 3.14159
try:
    radio = float(input("3. Ingresa el radio del círculo: "))
    # Usamos el operador ** para elevar al cuadrado
    area = pi * (radio ** 2)
    print(f"   El área del círculo es: {area:.2f}")
except ValueError:
    print("   Por favor, ingresa un valor numérico para el radio.")


# Ejercicio 4: Conversión
print("\n--- Ejercicio 4 ---")
numero_entero = 42
# Usamos str() para convertir el entero a una cadena
cadena_numero = str(numero_entero)
print(f"4. Tipo original: {type(numero_entero)}, Tipo convertido: {type(cadena_numero)}")
```

## 2. Estructuras de Control (Condicionales y Bucles)

```python
# EJERCICIOS 5-8: CONDICIONALES Y BUCLES

# Ejercicio 5: Condicional
print("--- Ejercicio 5 ---")
numero_prueba = -5
if numero_prueba > 0:
    print("5. El número es positivo.")
elif numero_prueba < 0:
    print("5. El número es negativo.")
else:
    print("5. El número es cero.")

# Ejercicio 6: Módulo (Par o Impar)
print("\n--- Ejercicio 6 ---")
num_par_impar = 13
# El operador módulo (%) devuelve el residuo. Si es 0, es par.
if num_par_impar % 2 == 0:
    print(f"6. El número {num_par_impar} es par.")
else:
    print(f"6. El número {num_par_impar} es impar.")

# Ejercicio 7: Bucle 'for'
print("\n--- Ejercicio 7 ---")
print("7. Números del 1 al 10:")
# range(inicio, fin+1)
for i in range(1, 11):
    print(i, end=" ")
print()

# Ejercicio 8: Bucle 'while'
print("\n--- Ejercicio 8 ---")
print("8. Bucle 'while' que se detiene con 0:")
entrada = -1 # Valor inicial diferente de 0
while entrada != 0:
    try:
        entrada = int(input("   Ingresa un número (0 para salir): "))
    except ValueError:
        print("   Entrada no válida. Intenta de nuevo.")
```

## 3. Colecciones (Listas, Diccionarios y Slicing)

```python
# EJERCICIOS 9-16: COLECCIONES

# Ejercicio 9: Crear una Lista
print("--- Ejercicio 9 ---")
frutas = ["manzana", "banana", "cereza", "dátil", "uva"]
print(f"9. Lista original: {frutas}")

# Ejercicio 10: Añadir a la Lista
print("\n--- Ejercicio 10 ---")
# Método .append() añade al final
frutas.append("kiwi")
print(f"10. Lista con kiwi: {frutas}")

# Ejercicio 11: Acceso por Índice
print("\n--- Ejercicio 11 ---")
# El segundo elemento está en el índice 1 (índices comienzan en 0)
segunda_fruta = frutas[1]
print(f"11. El segundo elemento es: {segunda_fruta}")

# Ejercicio 12: Crear un Diccionario
print("\n--- Ejercicio 12 ---")
estudiante = {
    "nombre": "Andrea Gómez",
    "edad": 22,
    "curso": "Desarrollo de Software"
}
print(f"12. Diccionario: {estudiante}")

# Ejercicio 13: Acceso a Diccionario
print("\n--- Ejercicio 13 ---")
# Acceso por la clave (key)
nombre_estudiante = estudiante["nombre"]
print(f"13. Nombre del estudiante: {nombre_estudiante}")

# Ejercicio 14: Máximo
print("\n--- Ejercicio 14 ---")
numeros = [45, 12, 88, 5, 91, 3]
# La función max() encuentra el valor más alto
maximo = max(numeros)
print(f"14. El máximo de {numeros} es: {maximo}")

# Ejercicio 15: Slicing
print("\n--- Ejercicio 15 ---")
# [inicio:fin] - Si omitimos el inicio, va desde el 0. El fin es exclusivo.
primeros_tres = frutas[:3]
print(f"15. Primeros 3 elementos de frutas: {primeros_tres}")

# Ejercicio 16: Listas Anidadas (Matriz 3x3)
print("\n--- Ejercicio 16 ---")
matriz_3x3 = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
print(f"16. Elemento central (fila 1, columna 1): {matriz_3x3[1][1]}")
```

## 4. Funciones

```python
# EJERCICIOS 17-20: FUNCIONES

# Ejercicio 17: Función simple
print("--- Ejercicio 17 ---")
def saludar_mundo():
    """Función que imprime un saludo simple."""
    print("17. ¡Hola Mundo!")

saludar_mundo()

# Ejercicio 18: Función con Parámetro
print("\n--- Ejercicio 18 ---")
def saludar_a(nombre):
    """Función que saluda a una persona específica."""
    print(f"18. ¡Hola, {nombre}! Bienvenido a Python.")

saludar_a("Fernando")

# Ejercicio 19: Función con Return
print("\n--- Ejercicio 19 ---")
def sumar_numeros(a, b):
    """Función que retorna la suma de dos números."""
    # El return es clave para devolver un valor
    return a + b

resultado = sumar_numeros(15, 7)
print(f"19. La suma de 15 y 7 es: {resultado}")

# Ejercicio 20: Función para Iterar
print("\n--- Ejercicio 20 ---")
def imprimir_lista(items):
    """Itera e imprime cada elemento de una lista."""
    print("20. Imprimiendo lista de ítems:")
    for item in items:
        print(f" - {item}")

imprimir_lista(["Lápiz", "Cuaderno", "Calculadora"])
```

## 5. Lógica, Strings y POO Básica

```python
# EJERCICIOS 21-27: LÓGICA, STRINGS Y POO

# Ejercicio 21: Strings (Count)
print("--- Ejercicio 21 ---")
texto = "programacion y desarrollo"
vocal = 'a'
conteo = texto.count(vocal)
print(f"21. La vocal '{vocal}' aparece {conteo} veces en '{texto}'.")

# Ejercicio 22: Strings (Upper)
print("\n--- Ejercicio 22 ---")
minusculas = "python es fundamental"
mayusculas = minusculas.upper()
print(f"22. En mayúsculas: {mayusculas}")

# Ejercicio 23: Lógica (Fibonacci)
print("\n--- Ejercicio 23 ---")
print("23. Serie de Fibonacci (10 elementos):")
a, b = 0, 1
for _ in range(10):
    print(a, end=" ")
    # El truco es que 'a' toma el valor de 'b', y 'b' toma la suma (a + b)
    a, b = b, a + b
print()

# Ejercicio 24: Lógica (Factorial)
print("\n--- Ejercicio 24 ---")
numero_fact = 5
factorial = 1
# Bucle que va de 1 hasta el número_fact (incluido)
for i in range(1, numero_fact + 1):
    factorial *= i
print(f"24. El factorial de {numero_fact} es: {factorial}")

# Ejercicio 25, 26, 27: POO (Clase, Constructor y Objeto)
print("\n--- Ejercicios 25-27: POO ---")
class Producto:
    """Clase base para un producto de inventario."""
    # Ejercicio 26: Constructor __init__
    def __init__(self, nombre, precio):
        """Método constructor para inicializar atributos."""
        self.nombre = nombre
        self.precio = precio
        
    def mostrar_info(self):
        return f"Producto: {self.nombre}, Precio: ${self.precio}"

# Ejercicio 27: Creación de instancias (objetos)
laptop = Producto("Laptop Gamer", 1500)
teclado = Producto("Teclado Mecánico", 80)
print(f"27. Objeto Laptop: {laptop.mostrar_info()}")
print(f"27. Objeto Teclado: {teclado.mostrar_info()}")
```

## 6. Tuplas y Manejo de Errores (Excepciones)

```python
# EJERCICIOS 28-30: TUPLAS Y EXCEPCIONES

# Ejercicio 28: Tuplas (Inmutables)
print("--- Ejercicio 28 ---")
# Son inmutables (no se pueden modificar) y se definen con paréntesis
coordenadas = (10, 20)
print(f"28. Tupla de coordenadas: {coordenadas}")

# Ejercicio 29: Excepciones (División por Cero)
print("\n--- Ejercicio 29 ---")
def dividir(a, b):
    # El bloque try intenta ejecutar el código
    try:
        resultado = a / b
        return resultado
    # El bloque except captura el error de división por cero
    except ZeroDivisionError:
        return "29. Error: No se puede dividir por cero."

print(f"División 10/2: {dividir(10, 2)}")
print(f"División 10/0: {dividir(10, 0)}")

# Ejercicio 30: Manejo de Archivos (Lectura)
print("\n--- Ejercicio 30 ---")

# Paso 1: Crear un archivo temporal (Modo 'w' de escritura)
with open("datos_prueba.txt", "w") as file:
    file.write("Línea 1 de datos\n")
    file.write("Línea 2 de datos\n")

# Paso 2: Abrir y leer el archivo (Modo 'r' de lectura)
print("30. Contenido del archivo datos_prueba.txt:")
try:
    # 'with open' asegura que el archivo se cierre después de leer
    with open("datos_prueba.txt", "r") as file:
        contenido = file.read()
        print(contenido)
except FileNotFoundError:
    print("El archivo no fue encontrado.")
```
